# BFS

```python
from queue import Queue

def bfs(startingNode, destinationNode):
    visited = {} 
    traversal_output = [] 
    frontier = Queue()

    for city in romanian_map.keys():
        visited[city] = False
        
    
    node = {'City':startingNode,'Distance':0, 'Path':[startingNode]}
    frontier.put(node)   
    
    while not frontier.empty():
        node = frontier.get()     
        u = node['City']
        visited[u] = True
        traversal_output.append(u)
        
        for v in romanian_map[u].keys():
            if not visited[v]:                
                d = node['Distance'] + romanian_map[u][v] 
                p =  node['Path'] + [v]
                frontier.put({'City':v,'Distance':d,'Path': p})
                if v == destinationNode: 
                    print(d)
                    print(traversal_output)
                    return p 
    print('not found')   
```

```python
bfs('Sibiu', 'Giurgiu')
```

```python

def dfs(startingNode, destinationNode):
    visited = {} 
    traversal_output = [] 
    Stack = [] 

    for city in romanian_map.keys():
        visited[city] = False
        
    
    node = {'City':startingNode,'Distance':0, 'Path':[startingNode]}
    Stack.append(node)   
    
    while  Stack:
        node = Stack.pop()     
        u = node['City']
        visited[u] = True
        traversal_output.append(u)
        
        for v in romanian_map[u].keys():
            if not visited[v]:                
                d = node['Distance'] + romanian_map[u][v] 
                p =  node['Path'] + [v]
                Stack.append({'City':v,'Distance':d,'Path': p})
                if v == destinationNode: 
                    print(d)
                    print(traversal_output)
                    return p 
    print('not found')
```
